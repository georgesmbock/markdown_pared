## Variables et types de données

En Python, on peut stocker des valeurs dans des variables.  
Exemple :
- `x = 5` (entier)
- `y = 3.14` (flottant)
- `nom = "Georges"` (chaîne de caractères)



```python
x = 5
y = 3.14
nom = "Georges"

print("x =", x)
print("y =", y)
print("nom =", nom)
```
